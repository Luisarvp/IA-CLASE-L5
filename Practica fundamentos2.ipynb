### 2. Imagenes como arreglos 

Una imagen no es mas que un arreglo estandar de Numpy que contiene pixeles de puntos de datos. Vuanto mayor sea el numero de pixeles en una imagen, mejor es su resolucion. Puede pensar que los pixeles son pequeños bloques de informacion dispuestos en forma de una cuadricula de 2D, y la profuncidad de un pixel se refiere a la informacion de color presente en ella. Para ser procesado por una computadora, una imagen debe convertirse en una forma binaria.

Numero de colores/sombras=2^bpp donde bpp represneta bits por pixel

Lo que obtenemos como resultado es un poco difernetes con respecto al color. Esperabamos una umagen de colores brillantes, pero lo que obtenemos es una imagen con cierto matiz azulado. Esto sucedo porque OpenCV y matplot.lib tienen diferentes ordenes de colores primarios. Mientras que OpenCV lee imagenes en forma de BGR, matplotlib, por otro lado, sigue el orden RGB. Por lo tanto, cuado leemos un archivo a traves de OpenCV, lo leemos como si contuviera canales en el orden azul, verde y rojo. Sin embargo, cuando mostramos la imagen usando matplotlib, el canalr rojo y el azul se intercambian y, por lo tanto, el tinte azul. Para evitar este problema transformaremos el canal a como matplotlib espera que use la funcion cvtcolor.

```python
img_rgb=cv2.cvtColor(img_raw, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

### Operaciones Basicas

### Dibujando en imagenes

Crea una imagen en negro que actuara como plantilla

```python
image_blank=np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
#Muestra la imagen en negro
plt.imshow(image_blank)
```

### Funcion y Atributos

La funcion general para dibujar formas o figuras en imagenes es:

##### CV2. Shape[line,rectangle,etc][image,pt1,pt2,color,thickness)

Hay alguns argumentos comunes que se pasan en funcion para dibujar figuras en las imagenes:

1) Imagen en la que se dibujaran las formas

2) Coordenadas de la figura a dibujar desde pt1 (arriba a la izquierda) a pt2(abajo a la derecha)

3) Color: el color de la forma que se va a dibujar. Se pasa como una tupia, por ejemplo:(255,0,0). Para la escala de grises, sera la escala de brillo.

4) El grosor de la figura geometrica.

#### a) Linea recta

Dibuja una linea recta a traves de una imagen requiere los puntos, a traves de los cuales pasara la linea


```python
# Dibuja una linea roja diagonal con un grosos de 5 pt
line_red=cv2.line(image_blank,(0,0),(511,511),(255,0,0),5)
plt.imshow(line_red)
```

```python
# Dibuja una linea verde diagonal con un grosos de 5 pt
line_green=cv2.line(image_blank,(0,0),(511,511),(0,255,0),5)
plt.imshow(line_red)
```

### Rectangulo

Para un rectangulo necesitamos especificar las coordenadas superior izquierda e inferior derecha

```python
# Dibuja un rectangulo azul con un grosos de 5 pt
line_red=cv2.rectangle(image_blank,(384,0),(500,128),(0,0,255),5)
plt.imshow(line_red)
```

### Escribir en imagenes

Agregar texto a las imagenes tambien es similar a dibujar formas en ellas. Pero debe especificar ciertos argumentos antes de hacerlo:

1) Escribir texto

2) Coordenadas de tecto. El texto en una imagen comienza desde la direccion inferior izquierda

3) Tipo de fuente y escala

4) Otros atributos como color, grosor, y tipo de linea. Normalmente, el tipo de linea que se usa es lineType=cv2.LINE_AA

```python
font=cv2.FONT_HERSHEY_SIMPLEX
text=cv2.putText(image_blank,'OpenCV',(10,300),font,4,(255,255,255),cv2.LINE_AA)
plt.imshow(text)
```
