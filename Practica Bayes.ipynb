Se utiliza clase SelectKBest de SKlearn para seleccionar las 5 mejores caracteristicas

```python
x=dataframe.drop(['comprar'],axis=1)
y=dataframe['comprar']

best=SelectKBest(k=5)
x_new=best.fit_transform(x,y)
x_new.shape

selected=best.get_support(indices=True)
print(x.columns[selected])
```

se visualiza la correlacion entre las 5 caracteristicas seleccionas

```python
used_features=x.columns[selected]

colormap=plt.cm.viridis
plt.figure(figsize=(12,12))
plt.title('Pearson Correlation of Feature', y=1.05, size=15)
sb.heatmap(dataframe[used_features].astype(float).corr(),linewidths=0.1,vmax=1.0,square=True, cmap=colormap, linecolor='white',annot=True)
```

### Ejemplo: Graficar 2 de las caracteristicas

```python
used_features2=[
    'ingresos',
    'financiar'
]
x=dataframe[used_features2].values
y=dataframe['comprar']
fig, ax=plt.subplots()
ax.scatter(x[:,0],x[:,1],c=y, s=50,cmap='RdBu')
ax.set_title('Naive Bayes Model',size=14)

xlim=(2000,9000)
ylim=(100000,700000)

xg=np.linspace(xlim[0],xlim[1],40)
yg=np.linspace(ylim[0],ylim[1],30)
xx, yy = np.meshgrid(xg,yg)
xgrid=np.vstack([xx.ravel(),yy.ravel()]).T

for label,color in enumerate (['red','blue']):
    mask=(y==label)
    mu, std =x[mask].mean(0), x[mask].std(0)
    p=np.exp(-0.*(xgrid-mu)**2/std).prod(1)
    pm=np.ma.masked_array(p,p<0.03)
    ax.pcolorfast(xg,yg,pm.reshape(xx.shape), alpha=0.5, cmap=color.title()+'s')
    ax.contour(xx,yy,p.reshape(xx.shape),levels=[0.01,0.05,0.09], colors=color, alpha=0.2)
ax.set(xlim=xlim, ylim=ylim)
```

Se aprecia que a mayor ingreso y una cantidad mayor de ingreso y una mantidad menor de financiamiento es mejor la opacion de comprar casa (en azul). De otra convendra rentar (en rojo)
